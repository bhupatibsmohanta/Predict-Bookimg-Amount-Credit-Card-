```python
import pandas as pd
import numpy as np

train=pd.read_excel('Credit_cards_App_data.xlsx')
x_train=train[['age','tdecision','empstaus','ExCus (Customer in Past)','Source','Salary','ExDebt (Liability)','Booking','Booking_Amt']]

print("No of Missing Values : \n",x_train.isna().sum())
x_train
```

```python
total_applications = x_train['tdecision'].count()
approved_applications = x_train['tdecision'][x_train['tdecision']=='Approve'].count()
print('Approval Rate : ', (approved_applications/total_applications))
```

```python
X=x_train[x_train['Booking'].notnull()]
client_decline=X[x_train['Booking']!='Y']
#print(client_decline['Booking'].describe())
client_decline['tdecision'].describe()

booked_applications = X['Booking'][x_train['Booking']=='Y'].count()
booked_applications
print('Booking Rate : ', booked_applications/approved_applications)
```

```python
#Non-Potential Customer
x_test = x_train[x_train['Booking'].isna()]

app_decline=x_test[x_test['tdecision']=='Decline']
app_decline

#Potential Customer
app_pending=x_test[x_test['tdecision']=='Pending']
app_pending
```

```python
x_train[x_train['tdecision']=='Approve']
#Since, Booking status is received from applicant after approval and assignment of Booking Amount
# So Booking amount doesn't depend on Booking Status. Let's drop the booking column
# as tdecision = 'Approve' also same for all rows let's drop it too. 
x_approved = x_train[x_train['tdecision']=='Approve'].drop(['Booking','tdecision'],axis=1)
app_pending = app_pending.drop(['Booking','tdecision'],axis=1)
x_approved
```

# Label Encoding for categorical data

```python
x_approved['empstaus'][x_approved['empstaus']=='NSE'] = 1
x_approved['empstaus'][x_approved['empstaus']=='SE'] = 0

x_approved['ExCus (Customer in Past)'][x_approved['ExCus (Customer in Past)']=='Y'] = 1
x_approved['ExCus (Customer in Past)'][x_approved['ExCus (Customer in Past)']=='N'] = 0

x_approved['Source'][x_approved['Source']=='EC']=1
x_approved['Source'][x_approved['Source']=='3P']=2
x_approved['Source'][x_approved['Source']=='D']=3
x_approved['Source'][x_approved['Source']=='MP']=4
x_approved
```

# Feature Scaling to improve data quality

```python
from sklearn.preprocessing import StandardScaler
age = StandardScaler()
salary = StandardScaler()
exdebt =StandardScaler()
amount = StandardScaler()
x_approved['age'] = age.fit_transform(x_approved['age'].values.reshape(699,1))
x_approved['Salary'] = salary.fit_transform(x_approved['Salary'].values.reshape(699,1))
x_approved['ExDebt (Liability)'] = exdebt.fit_transform(x_approved['ExDebt (Liability)'].values.reshape(699,1))
x_approved['Booking_Amt'] = amount.fit_transform(x_approved['Booking_Amt'].values.reshape(699,1))
```

```python
x_approved
```

# Support Vector Regression

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_approved.iloc[:,:6], x_approved.iloc[:,6], random_state=0,test_size=0.20)
```

```python
from sklearn.svm import SVR
reg = SVR(kernel = 'rbf')
reg.fit(X_train,y_train)
reg.score(X_test, y_test)
```

app_pending( prediction data) needs to undergo same transformation as training data.

```python
app_pending['age'] = age.transform(app_pending['age'].values.reshape(88,1))
app_pending['Salary'] = salary.transform(app_pending['Salary'].values.reshape(88,1))
app_pending['ExDebt (Liability)'] = exdebt.transform(app_pending['ExDebt (Liability)'].values.reshape(88,1))

app_pending['empstaus'][app_pending['empstaus']=='NSE'] = 1
app_pending['empstaus'][app_pending['empstaus']=='SE'] = 0

app_pending['ExCus (Customer in Past)'][app_pending['ExCus (Customer in Past)']=='Y'] = 1
app_pending['ExCus (Customer in Past)'][app_pending['ExCus (Customer in Past)']=='N'] = 0

app_pending['Source'][app_pending['Source']=='EC']=1
app_pending['Source'][app_pending['Source']=='3P']=2
app_pending['Source'][app_pending['Source']=='D']=3
app_pending['Source'][app_pending['Source']=='MP']=4
app_pending
```

# Prdiction of Booking Amount for potential customer

```python
y_pred = amount.inverse_transform( reg.predict(app_pending.iloc[:,:6]))

app_pending['age'] = age.inverse_transform(app_pending['age'].values.reshape(88,1))
app_pending['Salary'] = salary.inverse_transform(app_pending['Salary'].values.reshape(88,1))
app_pending['ExDebt (Liability)'] = exdebt.inverse_transform(app_pending['ExDebt (Liability)'].values.reshape(88,1))
app_pending['Booking_Amt']=y_pred.astype(int)
app_pending
```
